```python
#http://www.math.u-ryukyu.ac.jp/~suga/C/2004/8/node3.htmlpython
import math
import numpy as np
import matplotlib.pyplot as plt

# 
BOUNDS = 1.0  # 
SIZE = 500  # 
BASE = 255  # 
EPSILON = 1e-6
DUMMY = 0.0  # 

def hantei(x, y):
    u = DUMMY
    v = DUMMY
    nextu = x
    nextv = y

    if x == 0 and y == 0:
        return 0

    while cabs(nextu - u, nextv - v) > EPSILON:
        u = nextu
        v = nextv
        denom = 3 * (u**4 + 2 * u**2 * v**2 + v**4)
        nextu = (2.0 * u / 3.0) + (u**2 - v**2) / denom
        nextv = (2.0 * v / 3.0) - 2 * u * v / denom

    if abs(u - 1.0) < 0.5:
        return 1
    elif abs(v - (math.sqrt(3) / 2.0)) < 0.5:
        return 2
    elif abs(v + (math.sqrt(3) / 2.0)) < 0.5:
        return 3
    else:
        return 0

def cabs(a, b):
    return math.sqrt(a**2 + b**2)

def main():
    interval = 2 * BOUNDS / SIZE
    y = BOUNDS

    # RGB
    image = np.zeros((SIZE, SIZE, 3), dtype=np.uint8)

    for i in range(SIZE):
        x = -BOUNDS
        for j in range(SIZE):
            result = hantei(x, y)
            if result == 1:
                image[i, j] = [0, 0, BASE]  # 
            elif result == 2:
                image[i, j] = [0, BASE, 0]  # 
            elif result == 3:
                image[i, j] = [BASE, 0, 0]  # 
            else:
                image[i, j] = [0, 0, 0]  # 
            x += interval
        y -= interval

    # Matplotlib
    plt.imshow(image)
    plt.axis('off')  # 
    plt.show()

if __name__ == "__main__":
    main()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return x**4 - 3*x**3 + 2

def grad_f(x):
    return 4*x**3 - 9*x**2

def hess_f(x):
    return 12*x**2 - 18*x

def gradient_descent(start, lr, tol, max_iter):
    x = start
    path = [x]
    for _ in range(max_iter):
        grad = grad_f(x)
        x_new = x - lr * grad
        path.append(x_new)
        if abs(x_new - x) < tol:
            break
        x = x_new
    return x, path

def newton_method(start, tol, max_iter):
    x = start
    path = [x]
    for _ in range(max_iter):
        grad = grad_f(x)
        hess = hess_f(x)
        if hess == 0:
            break
        x_new = x - grad / hess
        path.append(x_new)
        if abs(x_new - x) < tol:
            break
        x = x_new
    return x, path

start = 2.5
learning_rate = 0.01
tolerance = 1e-6
max_iterations = 100

gd_min, gd_path = gradient_descent(start, learning_rate, tolerance, max_iterations)
newton_min, newton_path = newton_method(start, tolerance, max_iterations)

print(":", gd_min)
print(":", newton_min)

x_range = np.linspace(-1, 3, 500)
plt.plot(x_range, f(x_range), label="f(x) = x^4 - 3x^3 + 2", color="blue")
plt.scatter(gd_path, [f(x) for x in gd_path], color="red", label="Gradient Descent Path")
plt.scatter(newton_path, [f(x) for x in newton_path], color="green", label="Newton Method Path")
plt.title("Comparison of Gradient Descent and Newton Method")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.xlim(2.0, 3.0)
plt.ylim(-7.0, -5.5)
plt.legend()
plt.grid()
plt.show()
```
